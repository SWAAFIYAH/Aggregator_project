```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df_bookings = pd.read_csv('Datasets/Bookings.csv')
print(df_bookings)


#number of distict bookings
distinct_bookings = df_bookings['booking_id'].nunique()
print('Distinct bookings:')
print(distinct_bookings)







```

```python
df_session = pd.read_csv('Datasets/Sessions.csv') 
print(df_session)

#number of distict sessions
distinct_sessions = df_session['session_id'].nunique()
print('Distinct sessions:')
print(distinct_sessions)
```

```python
#number of distinct searches
distinct_search = df_session['search_id'].nunique()
print('Distinct searches:')
print(distinct_search)
```

```python
# Question2.How many sessions have more than one booking

#drop rows that do not have bookings
booked_sessions = df_session[df_session['booking_id'].notna()]
print('Sessions with bookings:')
print(booked_sessions)


#counting bookings per session

bookings_per_sessions = booked_sessions.groupby('session_id') ['booking_id'].count()

print('Bookings per session')
print(bookings_per_sessions)



#getting sessions with more than one booking

session_with_more_than_one_booking = bookings_per_sessions[bookings_per_sessions > 1]


print('\nSessions with more than one booking')
print(session_with_more_than_one_booking)







```

```python
# Question3. Which days of the week have the highest number of bookings? Also, draw a pie chart to show the distribution for all days of the week?

#converting booking time to booking date
df_bookings['booking_time'] = pd.to_datetime(df_bookings['booking_time'])


#Extracting booking day from booking date
df_bookings['day_of_week'] = df_bookings['booking_time'].dt.day_name()
print('booking days')
print(df_bookings['day_of_week'])


#counting booking per day
bookings_counts_per_day = df_bookings['day_of_week'].value_counts()

print('\nnumber of bookings for each day')
print(bookings_counts_per_day)

#getting the days with highest number of bookings
max_count = bookings_counts_per_day.max()

days_with_highest_bookings = bookings_counts_per_day[bookings_counts_per_day == max_count]
print('\ndays with highest number of bookings')
print(days_with_highest_bookings)

#pie chart to show distributions for all days of the week
# i have imported matplotlib.plt above

print('\nA pie chart showing the distribution of bookings per day')

plt.figure(figsize=(8,8))
bookings_counts_per_day.plot.pie(autopct='%1.1f%%')
plt.title('Booking Distributions per Day')
plt.show()



    
```

```python
#Question4 For each of the service names, display the total number of bookings and the total Gross Booking Value in INR.

#total bookings per service
total_booking_per_service = df_bookings.groupby('service_name').agg(
    total_bookings = ('booking_id', 'count')

)

print(total_booking_per_service)


#amount earned from the bookings in that service in INR
total_earnings_from_bookings = df_bookings.groupby('service_name').agg(
    total_earnings = ('INR_Amount', 'sum')
)

print('\nTotal earnings')
print(total_earnings_from_bookings)

```

```python
#question5 For customers who have more than 1 booking, which is the most booked route (from_city to to_city)? 


#customers booking count 
customers_booking_count = df_bookings.groupby('customer_id')['booking_id'].count()
print(customers_booking_count)


#customers with more than one booking
customers_with_more_than_one_booking = customers_booking_count[customers_booking_count > 1]

print("\nCustomers with more than one booking:")
print(customers_with_more_than_one_booking)



#most booked route city to city
most_booked_routes = df_bookings.groupby(['from_city', 'to_city']).size().reset_index(name='booking_count')
print(most_booked_routes)


# Get the route with the highest number of bookings
most_booked_route = most_booked_routes.sort_values(by='booking_count', ascending=False).head(1)

print("Most booked route:")
print(most_booked_route)


```

```python
#Question6 Which are the top 3 departure cities from where customers book mostly in advance, provided that there have been at least 5 departures from that city?


#group by depature city(from city) and count how many days earliar was the booking made and get the mean of the days for each booking
departure_city_statistics = df_bookings.groupby('from_city').agg(
    total_bookings = ('days_to_departure', 'count'),
    mean_for_the_days = ('days_to_departure', 'mean')
)

print(departure_city_statistics)


#top three cities
top_3_early_booking_cities = departure_city_statistics[
    departure_city_statistics['total_bookings'] >= 5
].sort_values(by='mean_for_the_days', ascending=False).head(3)

print('\ntop three cities')
print(top_3_early_booking_cities)
```

```python
#Question7 Plot a heatmap displaying correlations of the numerical column and report which pair of numerical columns in the bookings data set, have the maximum correlation?

#check which columns are numerical
numeric_dataframe = df_bookings.select_dtypes(include = ['number'])




#print correlation matrix
correlation_matrix = numeric_dataframe.corr()
print('\n correlation matrix')
print(correlation_matrix)



#plotting the heatmap

plt.figure(figsize = (5, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation heatmap')
plt.show()


# The heatmap shows that:
# INR_Amount and distance_km have a strong positive correlation (0.62).
# No_of_passengers and distance_km have no significant correlation (0.0017).
# INR_amount and days_to_departure have a weak  positive correlation(0.28)
#INR_amount and No_of_passengers have a a weak  positive correlation(0.4)
#days_to_departure and No_of_passengers have  a  very week  positive correlation(0.19)
#days_to_departure and distance_km have  a  very week  positive correlation(0.22)
```

```python
#Question8 For each service, which is the most used device type for making bookings on the platform? 

device_counts = df_bookings.groupby(['service_name', 'device_type_used']).size().reset_index(name='count')
print (device_counts)

#getting the row with the highest count for each service
most_used_devices = device_counts.loc[device_counts.groupby('service_name')['count'].idxmax()]
print('most used device')
print(most_used_devices)


```

```python
#Question9.Plot the trends at a quarterly frequency for the number of bookings by each of the device types, that is, plot a time series for each year and quarter showing the number of bookings performed by each device type. 

# Step 1: Convert 'booking_time' to datetime format
df_bookings['booking_date'] = pd.to_datetime(df_bookings['booking_time'])

# Step 2: Create a new 'quarter' column in the format '2023Q1', '2023Q2', etc.
df_bookings['quarter'] = df_bookings['booking_date'].dt.to_period('Q')

# Step 3: Group by quarter and device_type, and count the number of bookings
bookings_by_quarter = df_bookings.groupby(['quarter', 'device_type_used']).size().reset_index(name='num_bookings')

# Step 4: Pivot the table so each device_type becomes a column
pivot_table = bookings_by_quarter.pivot(index='quarter', columns='device_type_used', values='num_bookings').fillna(0)

# Step 5: Plot the data
pivot_table.plot(kind='line', marker='o', figsize=(12, 6))
plt.title("Quarterly Bookings by Device Type")
plt.xlabel("Year and Quarter")
plt.ylabel("Number of Bookings")
plt.grid(True)
plt.legend(title="Device Type")
plt.tight_layout()
plt.show()
```

```python
#Question10 12 customers performed a total of 100 searches but only a few of them performed 10 bookings in total. Searches = 100 bookings = 10 customers = 12 The overall booking to search ratio, Ober, is 10/100 = 0.1 Using the above information/context, answer the following: • What is the average oBSR for each month of the year? • What is the average oBSR for each day of the week? • Plot a time series of oBSR on all the given dates. 
# Convert search_time to datetime
df_session['search_time'] = pd.to_datetime(df_session['search_time'], format='mixed')

# Add a column to mark if this search led to a booking (1 = yes, 0 = no)
df_session['booked'] = df_session['booking_id'].notna().astype(int)


df_session['month'] = df_session['search_time'].dt.month

# Group by month and calculate mean of 'booked'
obsr_by_month = df_session.groupby('month')['booked'].mean()
print("Average oBSR per month:")
print(obsr_by_month)


df_session['weekday'] = df_session['search_time'].dt.day_name()

# Group by weekday and calculate mean of 'booked'
obsr_by_weekday = df_session.groupby('weekday')['booked'].mean()

# Sort days in order
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
obsr_by_weekday = obsr_by_weekday.reindex(days_order)

print("Average oBSR per weekday:")
print(obsr_by_weekday)


df_session['date'] = df_session['search_time'].dt.date

# Group by each date and calculate oBSR (mean of booked)
obsr_by_date = df_session.groupby('date')['booked'].mean()

# Plot
plt.figure(figsize=(12, 6))
obsr_by_date.plot()
plt.title("Time Series of oBSR by Date")
plt.xlabel("Date")
plt.ylabel("oBSR (Booking/Search Ratio)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
