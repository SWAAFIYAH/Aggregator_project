```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_bookings = pd.read_csv('Datasets/Bookings.csv')
print(df_bookings)


#number of distict bookings
distinct_bookings = df_bookings['booking_id'].nunique()
print('Distinct bookings:')
print(distinct_bookings)







```

```python
df_session = pd.read_csv('Datasets/Sessions.csv') 
print(df_session)

#number of distict sessions
distinct_sessions = df_session['session_id'].nunique()
print('Distinct sessions:')
print(distinct_sessions)
```

```python
#number of distinct searches
distinct_search = df_session['search_id'].nunique()
print('Distinct searches:')
print(distinct_search)
```

```python
# Question2.How many sessions have more than one booking

#drop rows that do not have bookings
booked_sessions = df_session[df_session['booking_id'].notna()]
print('Sessions with bookings:')
print(booked_sessions)


#counting bookings per session

bookings_per_sessions = booked_sessions.groupby('session_id') ['booking_id'].count()

print('Bookings per session')
print(bookings_per_sessions)



#getting sessions with more than one booking

session_with_more_than_one_booking = bookings_per_sessions[bookings_per_sessions > 1]


print('\nSessions with more than one booking')
print(session_with_more_than_one_booking)







```

```python
# Question3. Which days of the week have the highest number of bookings? Also, draw a pie chart to show the distribution for all days of the week?

#converting booking time to booking date
df_bookings['booking_time'] = pd.to_datetime(df_bookings['booking_time'])


#Extracting booking day from booking date
df_bookings['day_of_week'] = df_bookings['booking_time'].dt.day_name()
print('booking days')
print(df_bookings['day_of_week'])


#counting booking per day
bookings_counts_per_day = df_bookings['day_of_week'].value_counts()

print('\nnumber of bookings for each day')
print(bookings_counts_per_day)

#getting the days with highest number of bookings
max_count = bookings_counts_per_day.max()

days_with_highest_bookings = bookings_counts_per_day[bookings_counts_per_day == max_count]
print('\ndays with highest number of bookings')
print(days_with_highest_bookings)

#pie chart to show distributions for all days of the week
# i have imported matplotlib.plt above

print('\nA pie chart showing the distribution of bookings per day')

plt.figure(figsize=(8,8))
bookings_counts_per_day.plot.pie()
plt.title('Booking Distributions per Day')
plt.show()



    
```

```python
#Question4 For each of the service names, display the total number of bookings and the total Gross Booking Value in INR.

#total bookings per service
total_booking_per_service = df_bookings.groupby('service_name').agg(
    total_bookings = ('booking_id', 'count')

)

print(total_booking_per_service)


#amount earned from the bookings in that service in INR
total_earnings_from_bookings = df_bookings.groupby('service_name').agg(
    total_earnings = ('INR_Amount', 'sum')
)

print('\nTotal earnings')
print(total_earnings_from_bookings)

```

```python
#question5 For customers who have more than 1 booking, which is the most booked route (from_city to to_city)? 


#customers booking count 
customers_booking_count = df_bookings.groupby('customer_id')['booking_id'].count()
print(customers_booking_count)


#customers with more than one booking
customers_with_more_than_one_booking = customers_booking_count[customers_booking_count > 1]

print("\nCustomers with more than one booking:")
print(customers_with_more_than_one_booking)



#most booked route city to city
most_booked_routes = df_bookings.groupby(['from_city', 'to_city']).size().reset_index(name='booking_count')
print(most_booked_routes)


```
